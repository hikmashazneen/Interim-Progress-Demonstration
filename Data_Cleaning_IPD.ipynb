```python
import pandas as pd

# Columns to process
date_columns = [
    "IncidenceDate",
    "UpdatedDate",
    "TreatmentCompletionDate",
    "LastContactDate",
    "DateOfDeath"
]

# Store flagged entries
flagged_entries = []

def parse_date_flexible(value, row_idx, col_name):
    if pd.isna(value):
        return pd.NaT

    val_str = str(value).strip()

    # yyyymmdd
    if len(val_str) == 8 and val_str.isdigit():
        try:
            return pd.to_datetime(val_str, format="%Y%m%d")
        except:
            pass

    # dd/mm/yy convert to dd/mm/yyyy
    try:
        dt = pd.to_datetime(val_str, format="%d/%m/%y", errors='raise', dayfirst=True)
        return dt
    except:
        pass

    # general parsing (dd/mm/yyyy with single digits)
    try:
        return pd.to_datetime(val_str, dayfirst=True, errors='raise')
    except:
        flagged_entries.append((row_idx, col_name, val_str))
        return pd.NaT

# Process column by column
for col in date_columns:
    df[col] = [parse_date_flexible(val, idx, col) for idx, val in enumerate(df[col])]

# Display flagged entries
if flagged_entries:
    print("Flagged incomplete dates:")
    for row_idx, col_name, val in flagged_entries:
        print(f"Row {row_idx}, Column '{col_name}': {val}")
else:
    print("No incomplete entries found!")
```

# Noisy Data

```python
import pandas as pd
import numpy as np

def summarize_unique_values(df):
    summary = {}

    for col in df.columns:
        series = df[col].dropna()
        unique_vals = series.unique()
        n_unique = len(unique_vals)

        # If few unique values, list them safely
        if n_unique <= 20:
            summary[col] = {
                "unique_count": n_unique,
                "values": sorted(map(str, unique_vals))
            }

        else:
            # Pure numeric column
            if pd.api.types.is_numeric_dtype(series):
                summary[col] = {
                    "unique_count": n_unique,
                    "range": (series.min(), series.max())
                }

            # Datetime column
            elif pd.api.types.is_datetime64_any_dtype(series):
                summary[col] = {
                    "unique_count": n_unique,
                    "range": (series.min(), series.max())
                }

            # Mixed / object column
            else:
                summary[col] = {
                    "unique_count": n_unique,
                    "range": (
                        str(series.astype(str).min()),
                        str(series.astype(str).max())
                    )
                }

    return summary
```

```python
result = summarize_unique_values(df)

for col, info in result.items():
    print(f"\nColumn: {col}")
    print(info)
```

```python
df['Status'] = df.apply(lambda row: 1 if pd.notna(row['DateOfDeath']) or pd.notna(row['DiagnosisToDeathDuration']) else row['Status'], axis=1)
print(df)
```

```python
import pandas as pd
import numpy as np

#Convert columns to datetime objects
df['IncidenceDate'] = pd.to_datetime(df['IncidenceDate'])
df['UpdatedDate'] = pd.to_datetime(df['UpdatedDate'])

# Calculate months between
# Formula: (Year_Diff * 12) + Month_Diff
df['TimeInMonths'] = (df['UpdatedDate'].dt.year - df['IncidenceDate'].dt.year) * 12 + \
 (df['UpdatedDate'].dt.month - df['IncidenceDate'].dt.month)

# Identify and display records with a negative month difference
negative_diffs = df[df['TimeInMonths'] < 0]

print("\n--- Records with Negative Month Difference ---")
if not negative_diffs.empty:
    display(negative_diffs)
else:
    print("No negative differences found.")
```
